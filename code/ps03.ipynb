### 2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr
```

```python
def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

This is Allen's autocorrelation function from chap05 exercises. It computes the autocorrelation by finding the serial correlation over all the lag values. 

```python
lags, corrs = autocorr(bit_wave)
thinkplot.plot(lags, corrs, label='autocorr bit')
```

The autocorrelation function does decay quickly at the beginning, but it doesn't seem quite exponential. We see a spike at around 700 (index is shifted 700 and checked for serial correlation), which indicates that the signal at this lag is somewhat correlated to the original signal. 

### 3) Use autocorrelation to estimate the pitch of the saxophone recording in 100475__iluppai__saxophone-weep.wav.  What happens if you filter out the fundamental tone?  Is the estimate based on the autocorrelation function consistent with your perceived pitch?

```python
sax = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')
#sax.framerate
sax_wave = sax.segment(start=1, duration=0.8)
sax_wave.make_audio()
```

```python
lags2, corrs2 = autocorr(sax_wave)
thinkplot.plot(lags2, corrs2, label='autocorr sax')
thinkplot.show(axis=[0,500,-1,1])
```

This is the autocorrelation function for the second note in the saxophone weep wave. I had to zoom in because the frequencies in here are so high. The wave is periodic when lag is a little over 100, approximately 110 frames. The correlation is about 1 every 110 frames, so this corresponds to the fundamental frequency, or the lowest frequency in the signal.  

To figure out the fundamental frequency of this wave, we have to look at framerate/lag, which is 44100/110. Framerate is in frames per second and lag is in frames so we get 1/seconds from this calculation which is frequency.  

```python
44100/110 
```

It looks like the fundamental frequency is about 400 Hz. We can verify with the spectrum. 

```python
spec = sax_wave.make_spectrum()
thinkplot.config(xlim=[0,5000])
spec.plot()
```

Spectrum of the saxophone wave I just calculated the autocorrelation for. 

The lowest frequency here is close to 400, which shows you can find the fundamental frequency using autocorrelation. Now I'm going to try filtering out the fundamental frequency. 

```python
spec.band_stop(300,500)
thinkplot.config(xlim=[0,5000])
spec.plot()
```

Spectrum of wave with filtered out fundamental frequency

```python
new_sax = spec.make_wave()
new_sax.make_audio()
```

The pitch sounds the same even though I filtered out the fundamental frequency. Now I'm going to look at the autocorrelation function of this new wave. 

```python

lags2, corrs2 = autocorr(new_sax)
thinkplot.plot(lags2, corrs2, label='autocorr sax')
thinkplot.show(axis=[0,500,-1,1])
```

The autocorrelation function for the new saxophone wave with the fundamental frequency filtered out is the same as before, when the fundamental frequency wasn't filtered out. So the autocorrelation function is consistent with perceived pitch because even after you filter out the fundamental frequency, the perceived pitch stays the same. The autocorrelation function shows the perceived pitch to be at the fundamental frequency even when it's been filtered out.  

### 4) Test the algorithmic complexity of analyze1, analyze2 and scipy.fftpack.dct.  You will find code in ps03starter.ipynb that will help with this.

#### Complexity of analyze1

The following code is Allen's from the pso3starter code. I just look at the plots of ns versus time to look at how long it takes to compute the algorithm at different ns. I copy-pasted for each individual algorithm. 

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)

ns = 2 ** numpy.arange(4, 12)
print(ns)
```

```python
results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o dct.analyze1(ys, freqs, ts) #THIS IS THE ONLY THING I CHANGE TO LOOK AT EACH ALGORITHM
    #result = %timeit -o DCT(ys)
    results.append(result)
```

```python
bests = [result.best for result in results]
bests
```

```python
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
```

```python
x = numpy.log(ns)
y = numpy.log(bests)
t = scipy.stats.linregress(x,y)
t
```

It looks like for every order of magnitude ns increases, the time it takes to compute increases by 3 orders of magnitude when you use analyze1.

### 5) One of the major applications of the DCT is compression for both sound and images. In its simplest form, DCT-based compression works like this: Break a long signal into segments. Compute the DCT of each segment. Identify frequency components with amplitudes so low they are inaudible, and remove them. Store only the frequencies and amplitudes that remain. To play back the signal, load the frequencies and amplitudes for each segment and apply the inverse DCT. Implement a version of this algorithm and apply it to a recording of music or speech. How many components can you eliminate before the difference is perceptible?

```python
sax_wave = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')
sax_wave.make_audio()
```

```python
thinkplot.config(axis=[1,3,-11000,8000])
sax_wave.plot()
```

plot of sax wave

```python
full_wave = sax_wave.segment(start=0,duration=0)

def remove_not_strong_freqs(element):
    if abs(element)>10000000:
        return element
    else:
        return 0

for i in range(0,10,1):
    tinyseg = sax_wave.segment(start = i, duration = 1)
    tinyspec = tinyseg.make_spectrum()
    tinyspec.hs = map(remove_not_strong_freqs, tinyspec.hs)
    full_wave = full_wave.__or__(tinyspec.make_wave())

full_wave.make_audio()
    
```

```python
thinkplot.config(axis=[1,3,-11000,8000])
full_wave.plot()
```

plot of full_wave, which is sax_wave with any frequencies that don't have an amplitude highter than 10000000 are filtered out. 

Credits to Abi and Joshe for help with the map function and for reminding me that make_wave and make_spectrum were essentially the same as using inverse dct and dct. 

What the code does is take 10 seconds of the saxophone wave, divide it into small segments that are each 1 second long, and make spectrums for these small segments. I then modify the hs of these spectrums to remove any frequencies whose amplitudes are not above 10000000. Then, I turn the spectrums with the modified hs back to waves and piece them together. You can hear the difference in the two audio files where the frequencies that are not extremely prominent (don't have a high amplitude) are filtered out. This happens to result in filtering out a lot of the low notes. 

When I was removing frequencies with amplitudes less than a smaller number, like 10000, the sound didn't change that much.
