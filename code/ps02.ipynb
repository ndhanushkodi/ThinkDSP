If the duration is an integer multiple of the period, the beginning and end of the segment line up, and we get minimal leakage.

```python
spectrum = wave.make_spectrum()
spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

Here, rather than the original hamming window used in the example, I'll use another window.

```python
blackman_wave = wave
blackman_wave.ys = wave.ys*numpy.blackman(len(wave.ys))
blackman_spectrum = blackman_wave.make_spectrum()
blackman_spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

```python
bartlett_wave = wave
bartlett_wave.ys = wave.ys*numpy.bartlett(len(wave.ys))
bartlett_spectrum = bartlett_wave.make_spectrum()
bartlett_spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

Cool! Similar to the hamming window, the blackman window cuts out some of the leakage, but we lose even more total energy here than with the hamming window, though not by much. And in the bartlett window we lose even more energy, and there's a little more leakage than the blackman window. 

##### 2) Write a class called SawtoothChirp that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decreases) linearly.  Hint: combine the evaluate functions from Chirp and SawtoothSignal.  Plot the spectrogram of this signal. The effect of aliasing should be visible, and if you listen carefully, you can hear it.

```python
PI2 = numpy.pi*2
class SawtoothChirp(thinkdsp.Chirp):
    """Represents a sawtooth signal."""
    
    def _evaluate(self, ts, freqs):
        """Evaluates the signal at the given times.

        ts: float array of times
        
        returns: float wave array
        """
        dts = numpy.diff(ts)
        dphis = PI2 * freqs * dts
        phases = numpy.cumsum(dphis)
        
        cycles = phases / PI2
        frac, _ = numpy.modf(cycles)
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp)
        return ys

sawChirp = SawtoothChirp(start= 200, end= 1000)
sawWave = sawChirp.make_wave(duration = 2, framerate=11025)
sawWave.make_spectrogram(seg_length = 512).plot()
thinkplot.config(xlabel = 'time', ylabel = 'freq')
```

```python
sawWave.make_audio()
```

I can hear the aliasing- it sounds like someone is rubbing my ear while I listen to the sawtooth chirp sound. Also, the spectrogram shows that the fundamental frequency (the darkest blue line) increases from 200 to 1000Hz, like it's supposed to. And you can also see all the other frequencies (which drop off in the spectrum at a rate of 1/freq) as lighter blue lines. The lines crossing (with negative slopes) show the aliasing visually

##### 3)  In musical terminology, a “glissando” is a note that slides from one pitch to another, so it is similar to a chirp. A trombone player can play a glissando by extending the trombone slide while blowing continuously. As the slide extends, the total length of the tube gets longer, and the resulting pitch is inversely proportional to length.  Assuming that the player moves the slide at a constant speed, how does frequency vary with time? Is a trombone glissando more like a linear or exponential chirp?  Write a function that simulates a trombone glissando from C3 up to F3 and back down to C3. C3 is 262 Hz; F3 is 349 Hz.

```python
import math
class ExpoChirp(thinkdsp.Chirp):
    """Represents a signal with varying frequency."""
    
    def evaluate(self, ts):
        """Evaluates the signal at the given times.

        ts: float array of times
        
        returns: float wave array
        """
        start, end = math.log10(self.start), math.log10(self.end)
        freqs = numpy.logspace(start, end, len(ts)-1)
        return self._evaluate(ts, freqs)
```

```python
def glissando():
    trombone = ExpoChirp(start=262, end=349)
    trombone_wave = trombone.make_wave(duration=1.5)

    trombonedown = ExpoChirp(start=349, end=262)
    trombone_wavedown= trombonedown.make_wave(duration=1.5)
    trombonegliss = trombone_wave|trombone_wavedown
    return trombonegliss.make_audio()
    
glissando()
```

Since the length is increasing at a constant velocity, and pitch is inversely proportional to length, the pitch is increasing linearly. For the pitch to increase linearly, we need the frequency to increase exponentially, so the trombone glissando is similar to an exponential chirp. To make the glissando, we just need two expo chirps going from 262 to 349Hz and back from 349 to 262Hz.

##### 6) In a white noise signal, the mixture of frequencies changes over time. In the long run, we expect the power at all frequencies to be equal, but in any sample, the power at each frequency is random.  To estimate the long-term average power at each frequency, we can break a long signal into segments, compute the power spectral density for each segment, and then compute the average across the segments.   You can read more about this algorithm at http://en.wikipedia.org/wiki/Bartlett's_method.  Apply Bartlett's method to your natural noise source and see if it gives a clearer picture of the relationship between frequency and power.

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
wave = signal.make_wave(duration=1, framerate=32768)
specTOT = 0
for i in range(0,1,20):
    tinyseg = wave.segment(start = i, duration = 0.05)
    tinyspec = tinyseg.make_spectrum()
    specdens = tinyspec.hs**2/0.05
    specs = thinkdsp.Spectrum(hs=specdens, framerate=11025)
    specTOT+=specs

avg= specTOT.hs/20
thinkdsp.Spectrum(hs=avg, framerate=11025).plot()
thinkplot.show(xlabel = 'freq', ylabel='power density',xscale = 'log', yscale = 'log')
```

This is the generated power spectrum using the Bartlett method. I also plotted the power spectrum of the same UG noise below. They look pretty similar! It's cool to see that the slope here is 0, which is what corresponds to white noise. 

```python
spec = wave.make_spectrum()
spec.plot_power()
thinkplot.show(xlabel = 'freq', ylabel='power density',xscale = 'log', yscale = 'log')
```

In white noise like I just generated using the Bartlett method, we can see that the power is equal at all frequencies (to get that slope of 0).
