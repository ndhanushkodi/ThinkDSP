This is my implementation of the DFT. 

```python
i = complex(0, 1)
PI2 = 2 * math.pi
def DFT(ys):
    amps = []
    N = len(ys)
    for k in range(4):
        f=0
        for n in range(N):
            f += ys[n]*numpy.exp(-i*PI2*k*n/N)
        amps.append(f)
    return amps
            
DFT(wave2.ys)
```

To implement the DFT, I had to loop through the frequencies, so that for every frequency, you can compute the summation of ys[n]*e^(i*2pi*k*n/N). N is the total number of samples (so length of ys) and n loops through the samples in the summation. 

This is Allen's implementation for comparison. 

```python
def dft(ys):
    N = len(ys)
    ts = numpy.arange(N) / N
    freqs = numpy.arange(N)
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    amps = M.conj().transpose().dot(ys)
    return amps
dft(wave2.ys)
```

Our outputs looks very close!

3) In Chapter 7, I showed how we can express the DFT and inverse DFT as matrix multiplications. These operations are relatively fast, taking time proportional to N2, where N is the length of the wave array. That would be fast enough for many applications, but it turns out that there is a faster algorithm, the Fast Fourier Transform (FFT), which takes time proportional to N logN.

Read about the FFT at http://en.wikipedia.org/wiki/Cooley-Tukey_FFT_algorithm, and write an implementation. Hint: I suggest you write a simple version as a recursive function; don’t worry about “data reordering, bit reversal, and in-place algorithms”.

```python
def omega(N, k):
   return numpy.exp((PI2 * i * k) / N)

def fft(ys):
    N = len(ys)
    if N == 1:
        return ys
    else:
        Feven = fft([ys[i] for i in xrange(0, N, 2)])
        Fodd = fft([ys[i] for i in xrange(1, N, 2)])
 
        combined = [0] * N
        #print(combined)
        for k in xrange(int(N/2)):
            combined[k] = Feven[k] + omega(N, -m) * Fodd[k]
            #print(Feven)
            #print(Fodd)
            combined[k + int(N/2)] = Feven[k] - omega(N, -k) * Fodd[k]
 
        return combined
        
fft(wave2.ys)
```

The code implementation, for the most part, comes from this link Sophia sent us: http://jeremykun.com/2012/07/18/the-fast-fourier-transform/. It helped me understand, in the context of the variables used in the math, how to implement the algorithm. So basically I translated into variables I understood from the math on wikipedia. 

The Cooley-Tukey algorithm takes the ys array and recursively splits it into odd-indexed and even-indexed values and combines these two to make up the combined array of amplitudes that the fft returns. combined[k] combines the even and odd indexed values and combined[k+int(N/2)] replaces parts of the array that combined[k] creates. This part comes from the 'twiddle factor'.

#### Performance of autocorr from Chapter 5

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)

ns = 2 ** numpy.arange(4, 12)
print(ns)
```

```python
results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    w = thinkdsp.Wave(ys, framerate=11025)
    #result = %timeit -o dct.analyze1(ys, freqs, ts) 
    result = %timeit -o autocorr(w)
    results.append(result)
```

#### Performance of fft_autocorr from Chapter 8

```python
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

def fft_convolve(signal, window):
    fft_signal = numpy.fft.fft(signal)
    fft_window = numpy.fft.fft(window)
    return numpy.fft.ifft(fft_signal * fft_window)

def fft_autocorr(signal):
    N = len(signal)
    window = signal[::-1]
    signal = zero_pad(signal, 2*N)
    window = zero_pad(window, 2*N)

    corrs = fft_convolve(signal, window)
    corrs = corrs[N//2: 3*N//2]
    return corrs
```

```python
results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o fft_autocorr(ys)
    results.append(result)
```
